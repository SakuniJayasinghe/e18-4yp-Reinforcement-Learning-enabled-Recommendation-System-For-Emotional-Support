```python
import networkx as nx
import matplotlib.pyplot as plt
from collections import Counter
```

```python
ego_network = nx.Graph()
```

```python
def check_for_preferences(preferences_exist, preferences_new):
    count = 0
    
    count_1 = Counter(preferences_exist)
    count_2 = Counter(preferences_new)
    
    count = sum((count_1 & count_2).values())
    return count
```

```python
def add_node(name, preferences):
    global ego_network
    ego_network.add_node(name, Preferences=preferences)
    print('node has been added successfully')
    
    nodes = list(ego_network.nodes())
    print(f'node list is {nodes}')
    if len(nodes) > 1:
        for node in nodes:
            if node != name:
                node_pref = ego_network.nodes[node]['Preferences']
                if check_for_preferences(node_pref, preferences) > 0:
                    ego_network.add_edge(name, node)
```

```python
def get_complete_network():
    plt.figure(figsize=(8, 6))
    pos = nx.spring_layout(ego_network)  # You can choose a different layout algorithm if needed
    nx.draw(ego_network, pos, with_labels=True, font_weight='bold', node_size=800, node_color='skyblue')
    plt.title("Ego Network")
    plt.show()
```
